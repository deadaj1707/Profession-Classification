```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

# Load NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')



# Function to clean text
def clean_text(text):
    if isinstance(text, str):
        # Tokenize text
        tokens = word_tokenize(text.lower())

        # Remove stopwords and punctuation
        stop_words = set(stopwords.words('english'))
        tokens = [word for word in tokens if (word not in stop_words and word not in string.punctuation)]

        # Lemmatize words
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(word) for word in tokens]

        # Join tokens back into a string
        clean_text = ' '.join(tokens)

        return clean_text
    else:
        return ''
```

```python
# Clean the 'Text' column
acad_df['Text'] = acad_df['Text'].apply(clean_text)

# Assign each 'Category' as 2
acad_df['Category'] = 2

```

```python
# Clean the 'Text' column
law_df['Text'] = law_df['Text'].apply(clean_text)

# Assign each 'Category' as 0
law_df['Category'] = 0

# Display the cleaned DataFrame
print(law_df)
```

```python
# Clean the 'Text' column
cre_df['Text'] = cre_df['Text'].apply(clean_text)

# Assign each 'Category' as 4
cre_df['Category'] = 4

# Display the cleaned DataFrame
print(cre_df)
```

```python
# Clean the 'Text' column
acc_df['Text'] = acc_df['Text'].apply(clean_text)

# Assign each 'Category' as 1
acc_df['Category'] = 6

# Display the cleaned DataFrame
print(acc_df)
```

```python
combined_df = pd.concat([law_df, med_df, acad_df, eng_df, cre_df, mba_df, acc_df])
```

```python
# 1. Distribution of Categories
plt.figure(figsize=(8, 6))
sns.countplot(data=combined_df, x='Category')
plt.title('Distribution of Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
# 3. Word Cloud
word_counter = Counter(" ".join(combined_df['Text']).split())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_counter)
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Text Data')
plt.show()
```

```python
# 3. Keyword Analysis
from sklearn.feature_extraction.text import CountVectorizer

def get_top_keywords(texts, n=10):
    vectorizer = CountVectorizer(stop_words='english')
    X = vectorizer.fit_transform(texts)
    word_counts = X.sum(axis=0)
    word_freq = [(word, word_counts[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
    word_freq_sorted = sorted(word_freq, key=lambda x: x[1], reverse=True)
    return word_freq_sorted[:n]

top_keywords_by_category = {}
for category in combined_df['Category'].unique():
    texts_in_category = combined_df.loc[combined_df['Category'] == category, 'Text']
    top_keywords_by_category[category] = get_top_keywords(texts_in_category)

# Print top keywords by category
for category, top_keywords in top_keywords_by_category.items():
    print(f"Category {category} Top Keywords:")
    for keyword, count in top_keywords:
        print(f"{keyword}: {count}")
    print()
```
